#### Ques 9) write a Python program that asks the user for a number and determine whether it is positive negative or zero implement a loop that continues to ask the user for a number until they entre exit use break to exit the loop and continue to prompt for a new number if the input is not exit.

```python
while True:
    num=input("Enter Number: ")
    if num.lower() == 'exit':
        print("Exited!")
        break
    # Try to convert the input to a float
    try:
        num = int(num)
        # Determine if the number is positive, negative, or zero
        if num > 0:
            print(f"Number {num} is positive.")
        elif num < 0:
            print(f"Number {num} is negative.")
        else:
            print(f"Number {num} is zero.")  
    except ValueError:
        print("Invalid input!")
```

#### Ques 10) create a python script that takes two numbers as input and prints whether both numbers are even odd or one of each using relational and logical operators.

```python
# Take input from the user
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Check if both numbers are even
if num1 % 2 == 0 and num2 % 2 == 0:
    print("Both numbers are even.")
# Check if both numbers are odd
elif num1 % 2 != 0 and num2 % 2 != 0:
    print("Both numbers are odd.")
else:
    print("One number is even and one is odd.")
```

#### Ques 11) write a Python program that takes an integer Input and print its binary octal and hexadecimal equivalent using a for loop and bitwise operators.

```python
def convert_number(num):
    print(f"Decimal: {num}")
    
    # Binary
    print("Binary: ", end="")
    for i in range(31, -1, -1):
        print(1 if num & (1 << i) else 0, end="")
    print()
    
    # Octal
    print("Octal: {:o}".format(num))
    
    # Hexadecimal
    print("Hexadecimal: {:X}".format(num))

# Get user input
decimal_num = int(input("Enter a decimal number: "))

# Call the function to convert the number
convert_number(decimal_num)
```
